**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(d):
    y1998 = float(d['Average_Price'][d['Year']==1998])
    y2018 = float(d['Average_Price'][d['Year']==2018])
    ratio = y2018/y1998
    return ratio
```

```python
# check for borough 'Barking & Dagenham'
create_price_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham'])
```

```python
ratio_list = list()
for borough_name in dfg['London_Borough'].unique():
    ratio_list.append( (borough_name, create_price_ratio(dfg[dfg['London_Borough']==borough_name]) ) )
print(ratio_list)
```

```python
# Create DataFrame from this
df_ratios = pd.DataFrame(ratio_list,columns=('London_Borough','2018_Price_Ratio_to_1998')).sort_values('2018_Price_Ratio_to_1998',ascending=False).set_index(np.arange(1,len(ratio_list)+1))
df_ratios
```

```python
# Get top 15 boroughs with greatest change in price
top15 = df_ratios.head(15)
print(top15)
```

```python
# Plot
ax = top15[['London_Borough','2018_Price_Ratio_to_1998']].plot(kind='bar',figsize=(12,8))
ax.set_xticklabels(top15.London_Borough)
ax.set_title('Top 15 London Boroughs with Greatest Price Change')
ax.set_ylabel('Price Ratio from 1998 to 2018')
ax.set_xlabel('London Borough')
ax.get_legend().remove()
```

```python
# Create a data frame showing value increase as well as ratio
df_2018_1998 = dfg[dfg['Year']==2018].merge( dfg[dfg['Year']==1998], on='London_Borough', suffixes=('_2018','_1998') ).drop(columns=['Year_2018','Year_1998'])
df_2018_1998['Average_Price_Increase'] = df_2018_1998['Average_Price_2018'] - df_2018_1998['Average_Price_1998']
df_2018_1998.sort_values('Average_Price_Increase', ascending=False, inplace=True)
df_2018_1998.set_index(np.arange(1,len(df_2018_1998)+1), inplace=True)
df_2018_1998.reset_index(inplace=True)
df_2018_1998.rename(columns={'index':'Avg_Price_Increase_Rank'}, inplace=True)
df_2018_1998 = df_2018_1998.merge( df_ratios.reset_index().rename(columns={'index':'Price_Ratio_Rank'}), on='London_Borough' )

# Display top 15 price ratio
df_2018_1998.sort_values('Price_Ratio_Rank').head(15)
```
